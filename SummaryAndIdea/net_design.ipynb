共享权重是指在神经网络中，多个输入具有相同的权重参数。这意味着对于不同的输入情况，他们会经过同样的权重计算。

共享权重的意义在于可以提供一种参数共享的方式，使得网络能够有效地学习和泛化不同的输入情况。下面是共享权重的几个主要意义：

- 减少参数量：当输入具有相似的结构或特征时，通过共享权重可以减少网络的参数量。这样可以在有限的计算资源下提升网络的训练效率，并减少过拟合的风险。

- 提升模型泛化能力：共享权重可以使网络在不同的输入情况下学习到通用的特征表示。这样，即使在训练数据中没有涵盖到某些特定输入情况，网络也可以通过学习到的通用特征来进行预测和推理。

- 强化相关性：通过共享权重，网络可以将输入之间的相关性进行显式建模。这样，在进行预测任务时，网络可以更好地利用输入之间的相互作用，提高模型的表达能力和性能。


需要注意的是，共享权重并不适用于所有的网络结构和任务。这需要根据具体的问题和数据集进行判断和实验，并合理选择是否使用共享权重，并在实践中对其进行调优和验证。

## 使用策略和技术细节的关系

使用某个策略会带来数值计算时张量维度卯合要求不同（信息融合的不同），从而改变一个模型参数形状设计和前向计算算法的改变
